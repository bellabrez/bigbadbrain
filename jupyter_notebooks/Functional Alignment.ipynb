```python
def load_brain(file):
    brain = nib.load(file).get_data()
    brain = np.swapaxes(brain, 0, 1)
    brain = ants.from_numpy(brain)
    return brain

def rotate_brain(brain):
    if type(brain) is not np.ndarray:
        brain = brain.numpy()
    brain = np.swapaxes(brain, 0, 1)
    brain = ants.from_numpy(brain)
    return brain
```
