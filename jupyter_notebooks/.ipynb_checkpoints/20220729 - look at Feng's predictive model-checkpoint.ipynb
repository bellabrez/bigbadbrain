```python
import numpy as np
import os
import brainsss
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
directory = "/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20220624_supervoxels_in_FDA/20220729_feng_model/"

prediction_file = os.path.join(directory, 'most_expalined_neurons_real_adj_meta_out_list.npy')
predictions = np.load(prediction_file)

time_inidicies_file = os.path.join(directory, 'time_indices.npy')
time_inidicies = np.load(time_inidicies_file)
```

```python
time_inidicies.shape, predictions.shape
```

```python
####################
### Prep Fictrac ###
####################
func_path = '/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20190101_walking_dataset/fly_087/func_0/'
fictrac_path = os.path.join(func_path, 'fictrac')
fictrac_raw = brainsss.load_fictrac(fictrac_path)
timestamps = brainsss.load_timestamps(os.path.join(func_path, 'imaging'))

fps = 50
resolution = 10 #desired resolution in ms
expt_len = fictrac_raw.shape[0]/fps*1000
behaviors = ['dRotLabY', 'dRotLabZ']
fictrac = {}
for behavior in behaviors:
    if behavior == 'dRotLabY': short = 'Y'
    elif behavior == 'dRotLabZ': short = 'Z'
    fictrac[short] = brainsss.smooth_and_interp_fictrac(fictrac_raw, fps, resolution, expt_len, behavior, timestamps=timestamps,z=20)
fictrac_timestamps = np.arange(0,expt_len,resolution)
```

```python
time_inidicies.flatten().min()
```

```python
behavior_points = fictrac['Y'][time_inidicies.flatten()]
prediction_points = np.mean(predictions,axis=-1).flatten()
```

```python
plt.plot(fictrac['Y'])
```

```python
scipy.stats.pearsonr(behavior_points, prediction_points)
```

```python
plt.plot(behavior_points, prediction_points,'x')
```

```python
np.corrcoef(behavior_points, prediction_points)[0, 1]
```
