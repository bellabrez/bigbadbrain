```python
def load_FDA():
    FDA_file = '/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/anat_templates/20220301_luke_2_jfrc_affine.nii'
    FDA = np.asarray(nib.load(FDA_file).get_fdata().squeeze(), dtype='float32')
    FDA = ants.from_numpy(FDA)
    FDA.set_spacing((.38,.38,.38))
    FDA_lowres = ants.resample_image(FDA,(2.6,2.6,5),use_voxels=False)
    return FDA, FDA_lowres
```

```python
def load_synapses_in_FDA():
    hemibrain_mask_file = '/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20220624_supervoxels_in_FDA/synapses_transformed_FDA_0p38um_smoothed_mask.npy'
    hemibrain_mask = np.load(hemibrain_mask_file)
    return hemibrain_mask
```

```python
def get_hemibrain_bounding_box(mask):
    
    start = {}
    stop = {}
    for axis, name in zip([0,1,2],['x','y','z']):
        
        start[name] = np.min(np.where(mask != 0)[axis])
        start[name] = int(np.floor(start[name]))
        
        stop[name] = np.max(np.where(mask != 0)[axis])
        stop[name] = int(np.ceil(stop[name]))
        
    return start, stop
```

# Clean

```python
#############################
### GET VALID SUPERVOXELS ###
#############################

# get list of supervoxel ids that contain synapses
valid_clusters = list(np.where(np.sum(adj,axis=0) != 0)[0])
valid_clusters.remove(background_cluster)
```

```python
###################################################
### get distances between all VALID supervoxels ###
###################################################

centroids = []
for cluster_num in valid_clusters:
    xyz_voxel_space = np.asarray(np.where(labels_3d==cluster_num)).mean(axis=-1)
    centroids.append(list(xyz_voxel_space * (2.6,2.6,5)[0]))
centroids = np.asarray(centroids)

combs = []
distances = []
for comb in itertools.combinations(list(range(len(centroids))), 2):
    combs.append(comb)
    id1 = comb[0]; id2 = comb[1]
    
    x_dist = (centroids[id1,0] - centroids[id2,0])**2
    y_dist = (centroids[id1,1] - centroids[id2,1])**2
    z_dist = (centroids[id1,2] - centroids[id2,2])**2
    dist = (x_dist + y_dist + z_dist)**.5
    distances.append(dist)
distances = np.asarray(distances)
```

```python
distances.shape
```

```python
################################
### DEFINE "YES" SUPERVOXELS ###
################################

threshold = .2
#corrs_valid = ipsi_turn_corrs[np.asarray(valid_clusters)]
corrs_valid = fwd_corrs[np.asarray(valid_clusters)]
supervoxels_yes_valid = list(np.where(corrs_valid>threshold)[0])
print(len(supervoxels_yes_valid))
```

```python
#########################################################
### GET COMB PAIRS THAT CONTAIN SINGLE AND DOUBLE YES ###
#########################################################

# a really nice control will be to take each yes_supervoxel but sample from all supervoxels for edges
# so for a given distance range still pick supervoxels_no as usual, but have an additional filter for id
# I think the easiest way is to cut combs and distances down to only include pairs with at least 1 yes_supervoxel
# combs is currently for "valid" supervoxels, so lets convert yes_supervoxels to valid IDs

id1s = np.asarray([i[0] for i in combs])
id2s = np.asarray([i[1] for i in combs])

comb_id1_yes = np.asarray([False]*len(id1s))
comb_id2_yes = np.asarray([False]*len(id2s))
for i in supervoxels_yes_valid:
    comb_id1_yes[id1s == i] = True
    comb_id2_yes[id2s == i] = True

comb_id1_id2_yes = comb_id1_yes & comb_id2_yes
print(np.sum(comb_id1_id2_yes))
```

```python
#####################################################################
### DEFINE DISTANCE BINS AND GET NUMBER OF DOUBLE YES IN EACH BIN ###
#####################################################################

lowers = np.arange(0,150,10)
uppers = np.arange(10,160,10)

n_edges = []
for y in range(len(lowers)):
    in_distance_bin = ((distances > lowers[y]) & (distances < uppers[y]))
    in_distance_bin_and_is_double_yes = in_distance_bin & comb_id1_id2_yes
    bin_count = len(np.where(in_distance_bin_and_is_double_yes)[0])
    n_edges.append(bin_count)
print(n_edges)
```

```python
### BOOTSTRAP ###

bootstrap_num = 1000
connectivities = {}
for condition in ['all', 'singles', 'doubles']:
    connectivities[condition] = []
    
    #loop over distances
    for y in range(len(lowers)):

        supervoxels_all = ((distances > lowers[y]) & (distances < uppers[y]))
        if condition == 'all':
            supervoxel_subset = supervoxels_all
        elif condition == 'singles':
            supervoxel_subset = supervoxels_all & comb_id1_yes
        elif condition == 'doubles':
            supervoxel_subset = supervoxels_all & comb_id1_id2_yes

        # now I will want to take a specific number of these combos, randomly, n times
        labels_in_comb_space = np.where(supervoxel_subset)[0]
        sample_size = n_edges[y]

        connectivities_one_sample = []
        for n in range(bootstrap_num):
            one_sample_set = random.choices(labels_in_comb_space,k=sample_size)
            sample_connectivity = []
            for i in range(sample_size):
                id1 = combs[one_sample_set[i]][0]
                id2 = combs[one_sample_set[i]][1]
                id1 = valid_clusters[id1] # go back to adj space
                id2 = valid_clusters[id2]

                # correct for supervoxel size
                size = cluster_sizes[id1] + cluster_sizes[id2]

                sample_connectivity.append(adj[id1,id2]/size)
            connectivities_one_sample.append(np.mean(sample_connectivity))
        connectivities[condition].append(np.asarray(connectivities_one_sample))
```
