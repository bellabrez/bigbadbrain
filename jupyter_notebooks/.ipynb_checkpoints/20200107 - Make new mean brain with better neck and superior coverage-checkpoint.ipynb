# Load individual anatomy

```python
folder = os.path.join(root_directory,'20191204_anatomy_collection')
anats_to_load = ['f22a0.nii',
                 'f24a0.nii',
                 'f25a0.nii',
                 'f48a0.nii',
                 'f56a2.nii',
                 'f57a1.nii',
                 'f58a0.nii',
                 'f63a1.nii',
                 'f68a1.nii']
anatomy = {}
for this_anat in anats_to_load:
    anatomy[this_anat] = bbb.load_numpy_brain(os.path.join(folder, this_anat))
```

```python
# add a little padding to a close axis
anatomy['f24a0.nii']=np.append(np.zeros((20,512,251)),anatomy['f24a0.nii'],axis=0)
```

```python
anatomy_zscore = {}
for anat in anatomy:
    anatomy_zscore[anat] = (anatomy[anat]-np.mean(anatomy[anat]))/np.std(anatomy[anat])
    #print(np.max(to_mean[i,:,:,:]))
```

```python
anatomy_ants = {}
for anat in anatomy:
    fly_num = 'fly_{}'.format(anat[1:3])
    anat_num = 'anat_{}'.format(anat[4])
    anatomy_xml = os.path.join(root_directory, fly_num, anat_num, 'anatomy.xml')
    anatomy_ants[anat] = ants.from_numpy(anatomy_zscore[anat])
    anatomy_ants[anat].set_spacing(bbb.get_resolution(anatomy_xml))
    print('DONE: {}, {}'.format(anat, bbb.get_resolution(anatomy_xml)))
```

```python
for anat in anatomy:
    file = os.path.join(root_directory, '20191204_anatomy_collection', '20200107_collection', '{}.nii'.format(anat))
    ants.image_write(anatomy_ants[anat], file)
```

# Affine alignment to seed brain

```python
master = anatomy_ants['f24a0.nii']
for anat in anatomy_ants:
    print('Working on {}'.format(anat))
    t0 = time.time()
    slave = anatomy_ants[anat]
    slave2master = ants.registration(master,
                                     slave,
                                     type_of_transform='Affine')
    
    print('COMPLETE. Duration: {:0.0f} sec'.format(time.time()-t0))
    file = os.path.join(root_directory,
                        '20191204_anatomy_collection',
                        '20200107',
                        '{}.nii'.format(anat))
    ants.image_write(slave2master['warpedmovout'], file)
```

### Make them all the same range / bit depth

```python
anats_fixed = {}
for i,anat in enumerate(anatomy):
    if i in [0,1,2]:
        anats_fixed[anat] = anatomy[anat]*(2**8)/(2**13)
    else:
        anats_fixed[anat] = anatomy[anat]
    print(np.max(anatomy[anat]))
```

```python
for anat in anats_fixed:
    print(np.max(anats_fixed[anat]))
```

```python
anatomy_ants = {}
for anat in anats_fixed:
    fly_num = 'fly_{}'.format(anat[1:3])
    anat_num = 'anat_{}'.format(anat[4])
    anatomy_xml = os.path.join(root_directory, fly_num, anat_num, 'anatomy.xml')
    anatomy_ants[anat] = ants.from_numpy(anats_fixed[anat])
    anatomy_ants[anat].set_spacing(bbb.get_resolution(anatomy_xml))
    print('DONE: {}, {}'.format(anat, bbb.get_resolution(anatomy_xml)))
```
