```python
file = os.path.join(root_directory, 'stitched_brain_red.nii')
brain = np.asarray(nib.load(file).get_data(), dtype='float32')
```

```python
brain.shape
```

```python
plt.imshow(brain[:,:,100,5].T)
```

```python
meantime = np.mean(brain,axis=(0,1,2))
```

```python
plt.plot(meantime)
```

```python
brain_norm = brain/meantime
brain = None
```

```python
brain_norm.shape
```

```python
meanbrain = np.mean(brain_norm,axis=-1)
```

```python
error = []
for i in range(100):
    error.append(np.sum(np.abs(meanbrain - brain_norm[...,i])))
```

```python
plt.plot(error)
```

```python
plt.hist(error);
```

```python
good_frames = np.where(np.asarray(error)<3e7)[0]
print(len(good_frames))
```

```python
meanbrain_good = np.mean(brain_norm[...,good_frames],axis=-1)
```

```python
plt.figure(figsize=(20,10))
plt.imshow(meanbrain[300:800,100:300,100].T)
```

```python
plt.figure(figsize=(20,10))
plt.imshow(meanbrain_good[300:800,100:300,100].T)
```

```python
diff = abs(meanbrain - meanbrain_good)
```
