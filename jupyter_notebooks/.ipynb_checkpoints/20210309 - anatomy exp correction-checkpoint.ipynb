```python
## depth correction

for k,i in enumerate(range(126,225)):
    brain[:,:,i] += (brain[:,:,i] > 0)*(.00238*(k+1))
    
## clip off dim cell bodies
# .61 until 67
# .49 until 86
# .3 until 136 (or end)
brain[:,:,:67][brain[:,:,:67] < .61] = 0
brain[:,:,67:86][brain[:,:,67:86] < .49] = 0
brain[:,:,86:][brain[:,:,86:] < .3] = 0
```
