```python
def fft_signal(signal, sampling_rate, duration):
    """ Performs FFT on a signal.

    Parameters
    ----------
    signal: 1D numpy array
    sampling_rate: in Hz (I think... check)
    duration: int (in sec?). Can probably calulate this instead of requiring.

    Returns
    -------
    y: signal
    Y: fft
    t: times """
    
    Fs = sampling_rate
    y = signal
    Ts = 1.0/Fs; # sampling interval
    t = np.arange(0,duration,Ts) # time vector
    n = len(y) # length of the signal
    k = np.arange(n)
    T = n/Fs
    frq = k/T # two sides frequency range
    frq = frq[range(int(n/2))] # one side frequency range
    Y = np.fft.fft(y)/n # fft computing and normalization
    Y = Y[range(int(n/2))]
    return y, Y, t, frq
```

```python
def interp_brain(brain, resolution, z, expt_len):
    #resolution is in ms
    interp = interp1d(timestamps[:,z], brain[:,:,z,:], fill_value='extrapolate')
    xnew = np.arange(0,expt_len,resolution) #0 to last time at subsample res
    y = interp(xnew)
    return y
```

```python
def interp_fictrac(fictrac, fps, resolution, expt_len):
    #behavior = 'speed_all_3'
    sigma = 3
    camera_rate = 1/fps * 1000 # camera frame rate in ms
    
    x_original = np.arange(0,expt_len,camera_rate)
    dx = np.asarray(fictrac['dRotLabX'])
    dy = np.asarray(fictrac['dRotLabY'])
    dz = np.asarray(fictrac['dRotLabZ'])
    dx = scipy.ndimage.filters.gaussian_filter(dx,sigma=3)
    dy = scipy.ndimage.filters.gaussian_filter(dy,sigma=3)
    dz = scipy.ndimage.filters.gaussian_filter(dz,sigma=3)
    fictrac_smoothed = np.sqrt(dx*dx + dy*dy + dz*dz)
    #fictrac_smoothed = scipy.ndimage.filters.gaussian_filter(np.asarray(fictrac[behavior]),sigma=sigma)
    fictrac_smoothed = np.abs(fictrac_smoothed)
    #fictrac_smoothed = np.clip(fictrac_smoothed, a_min=0, a_max=None)
    fictrac_interp_temp = interp1d(x_original, fictrac_smoothed, bounds_error = False)
    xnew = np.arange(0,expt_len,resolution) #0 to last time at subsample res
    fictrac_interp = fictrac_interp_temp(xnew)

    # Replace Nans with zeros (for later code)
    np.nan_to_num(fictrac_interp, copy=False);
    
    return fictrac_interp
```

```python
resolution = 50 #desired resolution in ms
expt_len = 1000*30*60
fps = 50 #of fictrac camera
beta_len = 11
```

# Lets run PCA on multiple walking flies and compare

```python
expts = ['fly_25/func_0']
root_path = '/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20190101_walking_dataset'
```

```python
expt_len = 1000*30*60
fps = 50 #of fictrac camera

for expt in expts:
    # Load data
    print('Starting {}'.format(expt))
    directory = os.path.join(root_path, expt)
    pca_directory = os.path.join(directory, 'pca')
    try:
        os.mkdir(pca_directory)
    except:
        print('PCA directory already exists.')
    timestamps = bbb.load_timestamps(os.path.join(directory, 'imaging'))
    fictrac = bbb.load_fictrac(os.path.join(directory, 'fictrac'))
    brain_path = os.path.join(directory, 'brain_zscored_green.nii')
    brain = bbb.load_numpy_brain(brain_path)
    dims = bbb.get_dims(brain)

    # Interpolate fictrac
    fic = bbb.interpolate_fictrac(fictrac,timestamps,fps,expt_len,behavior='speed_all_3',sigma=3,sign=None)

    # Reshape brain for PCA
    X = brain.reshape(-1,brain.shape[-1])

    # Run PCA
    pca = PCA(n_components=10).fit(X)

    # Save components and variance
    file = os.path.join(pca_directory, 'components')
    np.save(file, pca.components_)
    file = os.path.join(pca_directory, 'explained_variance')
    np.save(file, pca.explained_variance_)

    # Cross-correlate with velocity and save figure
    plt.figure(figsize=(10,10))
    plt.title('Cross-Correlation of PCs (whole-brain) and Velocity')
    cors = []
    for i in range(10):
        corred = np.correlate(pca.components_[i,:],np.abs(fic[:,25]),mode='full')
        cors.append(corred) # for saving after loop
        plt.plot(bins[1:-1]/1000,corred,label='PC {}'.format(i+1))
    plt.xlim(-50,50)
    plt.axvline(0,linestyle='--',color='r')
    plt.xlabel('time, sec')
    plt.legend()
    file = os.path.join(pca_directory, 'PCs_behavior_crossCorr.png')
    plt.savefig(file,dpi=300,bbox_inches='tight')
    
    # Save CC
    file = os.path.join(pca_directory, 'cross_corr')
    np.save(file, np.asarray(cors))
    
    # Get spatial map and save
    scores = pca.transform(X)
    scores = np.reshape(scores, (dims['y'],dims['x'],dims['z'],10))
    file = os.path.join(pca_directory, 'spatial_PCs.nii')
    bbb.save_brain(file, scores)
```
