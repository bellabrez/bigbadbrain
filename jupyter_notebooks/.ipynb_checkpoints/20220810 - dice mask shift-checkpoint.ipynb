```python
def calc_dice(mask, neurons):
    intersect = np.logical_and(mask, neurons)
    dice = 2*np.sum(intersect,axis=(1,2,3))/(np.sum(neurons,axis=(1,2,3))+np.sum(mask))
    return dice
```
