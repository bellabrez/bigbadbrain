```python
root_directory = f'/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20190101_walking_dataset/'
```

```python
flies = ['087', '089', '092', '093', '094', '096',
         '097', '098', '099', '100', '101', '102', '105', '106', '110', '111']
```

```python
t0 = time.time()
aligned_brains = []
for fly in flies:
    mimic_file = os.path.join(root_directory, f'fly_{fly}/anat_0/moco/stitched_brain_green_mean.nii')
    moving_file = os.path.join(root_directory, f'fly_{fly}/anat_0/moco/anat_red_clean.nii')
    
    mimic = np.asarray(nib.load(mimic_file).get_fdata().squeeze(), dtype='float32')
    moving = np.asarray(nib.load(moving_file).get_fdata().squeeze(), dtype='float32')
    
    resolution = (.6,.6,1)

    print('downsampling')
    moving = ants.from_numpy(moving)
    moving.set_spacing(resolution)
    moving = ants.resample_image(moving,(2,2,2),use_voxels=False)
    
    mimic = ants.from_numpy(mimic)
    mimic.set_spacing(resolution)
    mimic = ants.resample_image(mimic,(2,2,2),use_voxels=False)
    
    print('aligning')
    moco = ants.registration(fixed, moving, type_of_transform='SyN')
    
    mimic_moco = ants.apply_transforms(fixed, mimic, moco['fwdtransforms'])
    aligned_brains.append(mimic_moco)
    print(time.time()-t0)
    #print(os.path.exists(mimic_file))
```
