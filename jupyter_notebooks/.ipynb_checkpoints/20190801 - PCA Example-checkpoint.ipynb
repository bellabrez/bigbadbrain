```python
import numpy as np
from time import time
import os
import sys
import matplotlib.pyplot as plt
import bigbadbrain as bbb
from sklearn.decomposition import PCA

%matplotlib inline
```

### Load brain data

```python
directory = '/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20190101_walking_dataset/fly_25/func_0'
brain_path = os.path.join(directory, 'brain_zscored_green.nii')
brain = bbb.load_numpy_brain(brain_path)
dims = bbb.get_dims(brain)
print('brain shape is x,y,z,t {}'.format(brain.shape))
```

### Use one slice. Reshape into voxels by time

```python
z=25
X = brain[:,:,z,:].reshape(-1,brain.shape[-1])
print('X is voxels by time {}'.format(X.shape))
```

### Perform PCA

##### https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

```python
pca = PCA().fit(X)
```

### I believe pca.components_ are the PCA "scores"?

```python
pca_scores = pca.components_
print('Scores is PC by time {}'.format(pca_scores.shape))
```

```python
plt.plot(pca_scores[0,:])
plt.title('PC1')
```

### I believe this is how I get the "loadings"?

```python
pca_loadings = pca.transform(X)
print('Loadings is voxel by PC {}'.format(pca_loadings.shape))
```

```python
pca_loadings_reshape = np.reshape(pca_loadings, (dims['y'],dims['x'],-1))
print('new shape: {}'.format(pca_loadings_reshape.shape))
```

```python
plt.imshow(pca_loadings_reshape[:,:,0].T)
plt.title('Voxel loadings')
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title('Cumulative variance explained')
plt.xlabel('PC')
plt.ylabel('Cumulative variance explained')
```
