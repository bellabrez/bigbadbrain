```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
import os
from scipy.interpolate import interp1d
from scipy.signal import correlate
from scipy.fftpack import fft,ifft

import warnings
warnings.filterwarnings('ignore')
```

## Load Neural Data (single voxel, 30,456 samples, aquired at 1.88Hz)

```python
load_file = '/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20210514_for_aws/Y.npy'
neural = np.load(load_file)
plt.figure(figsize=(10,10))
plt.subplot(211);
plt.plot(neural);
plt.title('Neural Data');
plt.subplot(212)
plt.hist(neural,bins=100);
plt.title('histogram');
```

## Load Behavior (forward velocity, 810,000 samples, aquired at 50Hz)

```python
load_file = '/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20210514_for_aws/20210525_all_beh.npy'
behavior = np.load(load_file)
behavior = np.clip(behavior,0,None)
behavior = np.nan_to_num(behavior)

plt.figure(figsize=(10,10))
plt.subplot(211);
plt.plot(behavior);
plt.title('Behavior');
plt.subplot(212)
plt.hist(behavior,bins=100);
plt.title('histogram');
```

## Load Behavior and Neural Timestamps; use to upsample neural to match behavior

```python
load_file = '/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20210514_for_aws/X_all_neural_timestamps_z_t.npy'
neural_timestamps = np.load(load_file)
neural_timestamps = neural_timestamps[:,29]
print(neural_timestamps.shape)
```

```python
load_file = '/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20210514_for_aws/20210525_all_beh_time.npy'
behavior_timestamps = np.load(load_file)
print(behavior_timestamps.shape)
```

```python
neural_interp_obj = interp1d(neural_timestamps, neural, bounds_error=False)
neural_upsampled = neural_interp_obj(behavior_timestamps)
neural_upsampled = np.nan_to_num(neural_upsampled)
print(neural_upsampled.shape)
```

## Calculate behavior auto-correlation

```python
num_tp = len(behavior)
half_tp = int(num_tp/2)
width = 250
```

```python
behavior_autocorr = scipy.signal.correlate(behavior, behavior, mode='same')
```

```python
plt.figure(figsize=(10,10))
plt.subplot(211)
plt.plot(behavior_autocorr);
plt.title('behavior auto-correlation; full');
plt.subplot(212)
plt.plot(behavior_autocorr);
plt.xlim(half_tp-width,half_tp+width)
plt.title('behavior auto-correlation; clipped to center +/- 5sec');
```

# Calculate behavior-neural cross-correlation

```python
behneu_crosscorr = scipy.signal.correlate(behavior, neural_upsampled, mode='same')
```

```python
plt.figure(figsize=(10,10))
plt.subplot(211)
plt.plot(behneu_crosscorr);
plt.title('behavior-neural cross-correlation; full');
plt.subplot(212)
plt.plot(behneu_crosscorr);
plt.xlim(half_tp-width,half_tp+width)
plt.title('behavior-neural cross-correlation; clipped to center +/- 5sec');
```

# Quick Tangent - comparing this cross-corr filter to an alternate approach: neural weighted average behavior. This method uses no interpolation of the neural data, rather for each neural datum the flanking behavior is averaged.

```python
load_file = '/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20210514_for_aws/X.npy'
behavior_windows = np.load(load_file)
print(f'behavior_windows is a matrix of behavior flanking each neural datum: {behavior_windows.shape}')
```

```python
neural.shape
```

```python
neuwebeh_filter = np.dot(behavior_windows.T, neural)
```

```python
plt.figure(figsize=(10,10))
plt.plot(behneu_crosscorr[half_tp-width:half_tp+width]/np.max(behneu_crosscorr), label='cross-corr')
plt.plot(neuwebeh_filter/10, label='neural-weighted-behavior')
plt.legend()
```

# Take Fourier Transforms

```python
### clip correlations to +/- 5sec window
behavior_autocorr_window = behavior_autocorr[half_tp-width:half_tp+width]
behneu_crosscorr_window = behneu_crosscorr[half_tp-width:half_tp+width]
```

```python
SAMPLE_RATE = 50
DURATION = 10 #sec
N = SAMPLE_RATE * DURATION
xf = scipy.fftpack.fftfreq(N, 1 / SAMPLE_RATE)
yf = fft(behavior_autocorr_window)

plt.figure(figsize=(10,10))
plt.suptitle('FFT of behavior-auto-correlation', fontsize=20)
plt.subplot(211)
plt.title('Output of fft')
plt.plot(yf)

plt.subplot(212)
plt.title('Output of fft, nicer visualization')
plt.plot(xf,np.abs(yf),'x')
plt.xlim(0,10)
plt.axvline(1/0.531995701,color='r',linestyle='--')
plt.xlabel('Hz')
plt.ylabel('power');
```

```python
SAMPLE_RATE = 50
DURATION = 10 #sec
N = SAMPLE_RATE * DURATION
xf = scipy.fftpack.fftfreq(N, 1 / SAMPLE_RATE)
yf = fft(behneu_crosscorr_window)

plt.figure(figsize=(10,10))
plt.suptitle('FFT of behavior-neural-cross-correlation', fontsize=20)
plt.subplot(211)
plt.title('Output of fft')
plt.plot(yf)

plt.subplot(212)
plt.title('Output of fft, nicer visualization')
plt.plot(xf,np.abs(yf),'x')
plt.xlim(0,10)
plt.axvline(1/0.531995701,color='r',linestyle='--')
plt.xlabel('Hz')
plt.ylabel('power');
```

# Calculate auto-corr via Fourier and individual samples

```python
window = 500 # this is 10 sec wide
beh_ = 0
for i in range(len(behavior)-window):
    beh = behavior[i:i+window]
    beh_ += fft(beh)*np.conj(fft(beh))
    
beh_ = ifft(beh_)
```

```python
plt.plot(beh_)
plt.title('auto-corr via DFT')
```

```python
window = 500 # this is 10 sec wide
cc = 0
for i in range(len(behavior)-window):
    beh = behavior[i:i+window]
    neu = neural_upsampled[i:i+window]
    cc += fft(neu)*np.conj(fft(beh))
    
cc = ifft(cc)
test = np.concatenate((cc[-250:], cc[:250]))
```

```python
plt.plot(test)
```

```python
window = 500 # this is 10 sec wide
top = 0
bot = 0
for i in range(len(behavior)-window):
    beh = behavior[i:i+window]
    neu = neural_upsampled[i:i+window]
    top += fft(neu)*np.conj(fft(beh))
    bot += fft(beh)*np.conj(fft(beh))
    
# cc = ifft(cc)
# test = np.concatenate((cc[-250:], cc[:250]))
```

```python
cc = ifft(top/(bot+10000))
test = np.concatenate((cc[-250:], cc[:250]))
plt.plot(test,'o')
```

```python
cc = ifft(top)
cc = np.concatenate((cc[-250:], cc[:250]))
plt.plot(cc/np.max(cc))

cc_norm = ifft(top/(bot+100000))
cc_norm = np.concatenate((cc_norm[-250:], cc_norm[:250]))
plt.plot(cc_norm/np.max(cc_norm))
```
