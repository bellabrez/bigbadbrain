# crop sama brains to make meanbrain

```python
dir_ = "/oak/stanford/groups/trc/data/berger/registration/meanbrain_building/sama_local_meanbrain_mirror/preprocessed/"
sama_brains = []
files = os.listdir(dir_)
file_names = []
for file in files:
    print(file)
    file_names.append(file)
    fullfile = os.path.join(dir_,file)
    temp = ants.image_read(fullfile)[:]
    sama_brains.append(temp)
```

```python
bbox = {'x_width': 500,
        'y_width': 300,
        'z_width': 50}

#xyz
centers = [(500,450,85),
           (580,450,90),
           (500,450,105),
           (550,450,80),
           (500,400,65),
           (580,350,90),
           (500,350,85)]

## crop
sama_brains_cropped = []
for i in range(len(sama_brains)):
    x_slice = slice(centers[i][0]-int(bbox['x_width']/2),centers[i][0]+int(bbox['x_width']/2))
    y_slice = slice(centers[i][1]-int(bbox['y_width']/2),centers[i][1]+int(bbox['y_width']/2))
    z_slice = slice(centers[i][2]-int(bbox['z_width']/2),centers[i][2]+int(bbox['z_width']/2))
    sama_brains_cropped.append(sama_brains[i][x_slice,y_slice,z_slice])
```

```python
dir_ = "/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20221223_sama_male_PB_crop"

for i in range(len(sama_brains)):
    save_file = os.path.join(dir_, file_names[i])
    nib.Nifti1Image(sama_brains_cropped[i], np.eye(4)).to_filename(save_file)
```

```python
plt.figure(figsize=(20,20))
for i in range(len(sama_brains_cropped)):
    plt.subplot(3,3,i+1)
    plt.imshow(sama_brains_cropped[i][:,:,25].T)
    plt.title(files[i])
```

```python
moving_2um_trimmed = moving_2um[10:-10,10:-10,3:]
moving_2um_trimmed = ants.from_numpy(moving_2um_trimmed)
moving_2um_trimmed.set_spacing((2,2,2))
```

```python
### width of subvol comes directly from subvol
# r="radius", o="original"
xr_o = int(np.ceil(moving_2um_trimmed.shape[0]/2))
yr_o = int(np.ceil(moving_2um_trimmed.shape[1]/2))
zr_o = int(np.ceil(moving_2um_trimmed.shape[2]/2))
print(xr_o,yr_o,zr_o)
```

```python
### estimate center of moving on FDA

xc = 158
yc = 46
zc = 16

plt.imshow(templates["FDA"][:,:,zc].T)
plt.axvline(xc,color='w')
plt.axhline(yc,color='w')
```

```python
### Mask FDA to the subvol

def newbox(c,r):
    return slice(*(np.clip(c-r,0,None),c+r))

new_box = []
for i,r in enumerate([xr_o,yr_o,zr_o]):
    r = r*2**0.5 # expand the box based on possible rotation
    new_box.append(int(r*moving_2um.spacing[i]/templates["FDA"].spacing[i]))
xr = new_box[0]
yr = new_box[1]
zr = new_box[2]

FDA_cropped = templates["FDA"][newbox(xc,xr),newbox(yc,yr),newbox(zc,zr)]
FDA_cropped = ants.from_numpy(FDA_cropped)
FDA_cropped.set_spacing(templates["FDA"].spacing)

FDA_padded = np.zeros((templates["FDA"].shape[0],templates["FDA"].shape[1],templates["FDA"].shape[2]))
FDA_padded[newbox(xc,xr),newbox(yc,yr),newbox(zc,zr)] = FDA_cropped.numpy()
FDA_padded = ants.from_numpy(FDA_padded)
FDA_padded.set_spacing(templates["FDA"].spacing)
```

```python
### align subvol ("moving") to padded FDA
subvol_to_lukeraw = ants.registration(FDA_padded, moving_2um_trimmed, type_of_transform='Affine')
```
