# Warp local to FDA and apply to indivs

```python
fixed_path = "/oak/stanford/groups/trc/data/Yukun/registration/mean_brain/FDA_downsampled_flip.nii"
fixed = ants.image_read(fixed_path)
fixed.set_spacing((.76,.76,.76))
fixed
```

```python
moving_path = "//oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20230130_LC11_full_analysis/meanbrain_building/clandinin/syn_6.nii"
moving = ants.image_read(moving_path)
moving.set_spacing((.65,.65,1))#(.49,.49,1))#(.76,.76,1))
moving
```

```python
warped = ants.registration(fixed,moving,'SyN')
```

```python
warped
```

```python
plt.imshow(warped["warpedmovout"][:,:,100].T)
```

```python
plt.imshow(fixed[:,:,100].T)
```

```python
save_file = "/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20230130_LC11_full_analysis/local_to_FDA/clandinin_local2FDA"
nib.Nifti1Image(warped["warpedmovout"].numpy(), np.eye(4)).to_filename(save_file)
```

```python
save_path = "/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20230130_LC11_full_analysis/local_to_FDA/murthy"
os.mkdir(save_path)
```

```python
mimic_path = "/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20230130_LC11_full_analysis/indiv_to_local/clandinin"
save_path = "/oak/stanford/groups/trc/data/Brezovec/2P_Imaging/20230130_LC11_full_analysis/local_to_FDA/clandinin"

for file in os.listdir(mimic_path):
    print(file)
    mimic = ants.image_read(os.path.join(mimic_path, file))
    mimic.set_spacing((.65,.65,1))#(.49,.49,1))#(.76,.76,1))
    moco = ants.apply_transforms(fixed,mimic,warped['fwdtransforms'])
    save_file = os.path.join(save_path,file)
    nib.Nifti1Image(moco.numpy(), np.eye(4)).to_filename(save_file)
```
